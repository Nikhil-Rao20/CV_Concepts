# Histogram of Gradients : HOG

## Using library functions

### Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage.feature import hog
from skimage import io, color
```

### Loading the Image

```python
image_path = '2.jpg'
```

```python
image = io.imread(image_path)  
```

### Converting it into gray scale

```python
gray_image = color.rgb2gray(image)
```

### From using the hog function from skimage library to extract the features and image

```python
hog_features, hog_image = hog(gray_image, pixels_per_cell=(8, 8), cells_per_block=(2, 2), visualize=True)
```

### Plottng the HOG-Image

```python
plt.figure(figsize=(8, 8))
plt.imshow(hog_image, cmap='RdYlBu_r')
plt.title("HOG Features")
plt.show()
```
